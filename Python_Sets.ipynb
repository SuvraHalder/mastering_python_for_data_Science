# Sets  

A set is an unordered collecton of items . Every set element is unique and must be immutable .
Sets can also be used to perform mathematical set operations like union , intersection , symmetric diffrence , etc.

### Characteristics:

- Unordered
- Mutable
- No Duplicates
- Cant contain mutable data types


# Creating Sets

```python
# Empty Set
s = {}
print(s)
print(type(s))
```

```python
s  = set()
print(s)
print(type(s))
```

```python
# 1D AND 2D
s1 = {1,2,3}
print(s1)

# 2D SETS ARE NOTPOSSIBLE  BECAUSE SET ELEMETS SHOULD BE IMMUTABLE BUT
# SET IS ITSELF MUTABLE .
s2 = { 3 , 4 , s1 }
```

```python
# heterogeneous set

s3 = { 1 ,"hello" , 4.5 , True , (1,2,3)}
print(s3)

# homogenous set

s3 = {1,2,3,4,5}
print(s3)
```

```python
#using Type Conversiom
s4  = set([1,2,3,4])
print(s4)
```

```python
# duplicates not allowed
s5={1,1,2,2,3,3}
print(s5)
```

```python
#set Cant have mutable items
s6 = {1,2,3,4,[1,3,4,5]}
```

#Accessing and Editing Items

 - Set items Cant be Aceesed nor edited  because  Set is unodered

#Adding Items

```python
s = {1,2,3,4}

#add
s.add(5)

print(s)

#update -> we can use multiple items

s.update([5,6,7])
print(s)
```

# Deleting Items


```python
# Using del keyword we can delete the set but cant delete any item .
del  s
```

```python
# discard

s = {1,2,3,4,5}
s.discard(5)
print(s)

s = {1,2,3,4,5}

#remove
s.remove(5)
print(s)

'''
   discard and remove are similar  but if we try to delete an element
   which is not set ,
   discard -> will not throw Exception.
   remove -> will throw Exception .

'''

# pop -> Randomly deletes any item we have 0 control

s.pop()
print(s)

# clear -> clears whole set and makes it empty
s.clear()
print(s)


```

# Set Operations

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# Union -> Basic Set math union( | )
print(s1.union(s2))
print(s1 | s2)

#Intersection ( & ) -> simple intersection of sets
print(s1 & s2)

# Normal Diffrence
print(s1 - s2)

# Symmetric Diffrence ( ^ )
print( s1  ^ s2 )

# Memebership Test
print(1 in s1)

#interation
for i in  s1:
  print(i ,end=" ")

```

# Set Functions

```python
#len/sum/min/max/sorted

s = {1,2,3,4,5}
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(len(s))  # -> Gives You Length of Set

print(sum(s)) # -> gives You Sum

print(min(s),max(s))

print(sorted(s))  #-> Sorted always return List

print(s1.union(s2)) # same as  Union Operator but its a function

print(s1.update(s2)) # ->  does union and adds to s1

print(s1.intersection(s2)) # same as & Operator but its a function

s1.intersection_update(s2) # intersection result  stores at s1
print(s1)

```

#Symmetric_diffrence

A Δ B = (A - B) ∪ (B -A)

```python
s1 = {1,2,3,4,5}
s2 ={4,5,6,7,8}

s1.difference(s2)
s1.difference_update(s2) #does the diffrence and add the result to s1

s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2) # does symmetric diffrence and add the result to s1
```
